```python
class Solution(object):
    def longestPalindrome(self, s):
        """
        :type s: str
        :rtype: str
        """
        def reverse_string(s):
            return ''.join(reversed(s))
        
        def is_palindrome(s):
            return s.lower() == reverse_string(s).lower()
                
        def determine_longest_palindrome(candidate_list):
            if not candidate_list:
                return ''
            
            lengths_of_longest_seen_palindrome_candidates = map(len, candidate_list)
            maximum_length_of_longest_seen_palindrome_candidates = max(lengths_of_longest_seen_palindrome_candidates)
            index_of_lengths_of_longest_seen_palindrome_candidates = lengths_of_longest_seen_palindrome_candidates.index(maximum_length_of_longest_seen_palindrome_candidates)
            return candidate_list[index_of_lengths_of_longest_seen_palindrome_candidates]            
        
        def find_palindromes(s):
            if is_palindrome(s):
                return s
            else:
                find_left = find_palindromes(s[:-1])
                find_right= find_palindromes(s[1:])
                
                if len(find_left) > len(find_right):
                    return find_left
                else:
                    return find_right

        return find_palindromes(s)
```
